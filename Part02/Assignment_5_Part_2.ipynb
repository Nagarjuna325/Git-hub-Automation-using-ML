```python
import networkx as nx

# Create an empty graph
graph = nx.Graph()

# Add nodes for each object and property
graph.add_nodes_from(['Topic', 'Tag', 'Repository', 'Issue',
                      'Topic name', 'Tag name', 'Repository name', 
                      'Repository description', 'Issue title', 'Issue description', 
                      'Issue Id' , 'Issue Number'])

# Add links between the objects
graph.add_edges_from([('Repository', 'Topic'),
                      ('Topic','Tag'),
                      ('Repository', 'Tag'),
                      ('Repository', 'Issue'),
                      ('Issue', 'Tag'), 
                      ('Topic', 'Topic name'),
                      ('Tag','Tag name'),
                      ('Repository', 'Repository name'),
                      ('Repository', 'Repository description'),
                      ('Issue','Issue title'),
                      ('Issue','Issue description'),
                      ('Issue','Issue Id'),
                      ('Issue','Issue Number'),])

# Visualize the graph
nx.draw_networkx(graph, with_labels=True)

```

### Requirement 7: 


```python
#Add your code for this requirement in this cell
import requests

# Function to retrieve data from GitHub
def get_github_data(repo_name):
    # Construct the GitHub API URL
    url = f"https://api.github.com/repos/{repo_name}"
    
    # Make a GET request to the GitHub API
    response = requests.get(url)
    
    # Parse the JSON response and return the stargazers count
    data = response.json()
    return data
    #print(f"items: {data['items']}")

# Function to retrieve data from StackOverflow
def get_stackoverflow_data(tag_name):
    # Construct the StackExchange API URL
    url = f"https://api.stackexchange.com/2.3/questions?order=desc&sort=activity&tagged={tag_name}&site=stackoverflow"
    
    # Make a GET request to the StackExchange API
    response = requests.get(url)
    
    # Parse the JSON response and return the total number of questions
    data = response.json()
    return data

# URLs to retrieve data for
urls = [
    {
        "name": "React",
        "github_url": "https://github.com/facebook/react",
        "stackoverflow_tag": "reactjs"
    },
    {
        "name": "Angular",
        "github_url": "https://github.com/angular/angular",
        "stackoverflow_tag": "angular"
    },
    {
        "name": "Visualization",
        "github_url": "https://github.com/d3/d3",
        "stackoverflow_tag": "d3.js"
    }
]

# Loop through each URL and retrieve the data
for url in urls:
    # Retrieve data from GitHub
    github_data = get_github_data(url["github_url"].split('item')[-1])
    
    # Retrieve data from StackOverflow
    stackoverflow_data = get_stackoverflow_data(url["stackoverflow_tag"])
    
    # Print the data
    print(f"{url['name']}: GitHub Stars - {github_data}, StackOverflow Questions - {stackoverflow_data}")

```

```python
#Add your code for this requirement in this cell

import requests
import pandas as pd

# Set up the API endpoint and parameters
api_url = 'https://api.stackexchange.com/2.3/questions'
params = {'site': 'stackoverflow', 'tagged': 'python', 'pagesize': 100}

# Make the API request and get the data
response = requests.get(api_url, params=params)
data = response.json()['items']

# Create a DataFrame to store the data
df = pd.DataFrame(data)

# Print out the number of questions retrieved
print(f'Number of questions retrieved: {len(df)}')

# Group the questions by tag and count the number of questions for each tag
tag_counts = df.groupby('tags').size()

# Print out the tag with the highest number of questions
max_tag = tag_counts.idxmax()
print(f'The tag with the highest number of questions is {max_tag}, with {tag_counts[max_tag]} questions.')
```
